# Handling missing values

```python
# We observed that Rating column has 1474 missing values.
# so we will remove those values from our data set.
```

```python
# Check the missing values in Rating column.
data.Rating.isnull()
```

```python
# Dropping the missing rows in Rating column.
data1=data[-data.Rating.isnull()]
data1.head()
```

```python
# Check the size of both of the data frames
print(data.shape)
print(data1.shape)
```

```python
# Again check for missing value in the dataframe
data1.isnull().sum()
```

```python
# Check for null values in Current ver column.
data1[data1["Current Ver"].isnull()]
```

```python
# Since only 4 values are missing, so we will replace the missing values with the most comman Current ver .
```

```python
# Check for most comman current version
data1["Current Ver"].value_counts()
```

```python
# We observed that "Varies with device" is the most cooman current version.
# so we will replace missing values with Varies with device.
```

```python
data1["Current Ver"]=data1["Current Ver"].fillna(data1["Current Ver"].mode()[0])
```

```python
data1[data1["Current Ver"].isnull()]
```

```python
# We have filled all the missing values in the Current ver column.
data1.shape
```

```python
# Again check for missing values
data1.isnull().sum()
```

```python
# We are checking for missing value in content rating column.
data1[data1["Content Rating"].isnull()]
```

```python
# Removing the missing row .
data2=data1[-data1["Content Rating"].isnull()]
data2.shape
```

```python
# Checking for null values again.
data2.isnull().sum()
```

```python
# Checking for null values in Android version column.
data2[data2["Android Ver"].isnull()]
```

```python
# Checking for most popular Android version.
data2["Android Ver"].value_counts()
```

```python
# Replace the missing values with "4.1 and up".
data2["Android Ver"]=data2["Android Ver"].fillna(data2["Android Ver"].mode()[0])
```

```python
# Checking for missing values again in the dataframe.
data2.isnull().sum()
```

```python
# Observing the shape of the dataframes.
print(data.shape)
print(data1.shape)
print(data2.shape)
```

```python
# SO far we have handled the missing values from the dataset.
```

# Handling incorrect datatypes

```python
# Check for datatypes of the columns.
data2.info()
```

```python
# Observing price column.
data2["Price"].value_counts()
```

```python
# Since the price sould be in numerical format , so we are making the necessary changes.
```

```python
data2["Price"]=data2["Price"].apply(lambda x: 0 if x=="0" else float(x[1:]) )
```

```python
data2.Price.value_counts()
```

```python
# Checking the datatype again.
data2.info()
```

```python
# Checking the reviews column
data2.Reviews.value_counts()
```

```python
# Changing the datatype of Reviews column.
```

```python
data2.Reviews=data2.Reviews.astype("int32")
```

```python
data2.info()
```

```python
# Making changes in Installs column.
data2["Installs"]=data2["Installs"].apply(lambda x : int(x.replace(',','').replace('+','')))
```

```python
data2["Installs"]
```

```python
# Checking datatype once again
data2.info()
```
